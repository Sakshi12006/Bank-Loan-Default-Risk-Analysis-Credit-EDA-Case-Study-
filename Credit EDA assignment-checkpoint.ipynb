### finding numerical columns and null values in numerical columns

```python
numerical_cols=[cols for cols in df1.columns if df1[cols].dtypes=='int64' or df1[cols].dtypes=='float64']
numerical_cols
```

```python
null_counts= df1[numerical_cols].isnull().sum()
null_counts
```

```python
null_percentage= (null_counts/len(df1))*100
sorted_null_percentage= null_percentage.sort_values(ascending=False)
sorted_null_percentage
```

#### we can remove/drop the columns which has more than 40% of null values 

```python
removed_columns= null_percentage[null_percentage>40].index
df1.drop(removed_columns, axis=1, inplace=True)
```

```python
df1.shape
```
